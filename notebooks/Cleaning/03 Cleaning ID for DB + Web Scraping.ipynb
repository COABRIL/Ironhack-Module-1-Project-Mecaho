# Minimun cleaning for merge

```python
dfpop = dfpop.drop(dfpop.index[240])
```

```python
def rename_c(df, old, new):
    df.rename(columns={old: new}, inplace=True)
    return df

rename_c(dfpop, 'Country(or dependent territory)', 'country')
rename_c(dfpop, '% of World Population', 'weight_wpopulation')
rename_c(dfpop, 'Rank', 'rank_pop')
rename_c(dfpop, 'Population', 'population')
rename_c(dfpop, 'Date', 'date')
```

```python
def del_str_patt(df, col, del_str, new_str):
    df[col] = df[col].apply(lambda x: re.sub(del_str, new_str, str(x))).str.strip()
    return df

del_str_patt(dfpop, "country", '\([^()]*\)', '')
del_str_patt(dfpop, "country", '\[[^\]]*\]', '')
```

```python
def replace(df, col, old_v, new_v):
    df[col] = df[col].str.strip().replace(old_v, new_v)
    return df

replace(dfforb, 'country', 'USA', 'United States')
replace(dfforb, 'country', "People's Republic of China", 'China')
replace(dfforb, 'country', 'UK', 'United Kingdom')
replace(dfforb, 'country', 'Swaziland', 'Eswatini')
```
