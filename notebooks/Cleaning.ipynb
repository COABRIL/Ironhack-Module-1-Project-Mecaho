# Import libraries

```python
import pandas as pd
import re
import numpy as np
```

# Read csv

```python
df = pd.read_csv('../../Ironhack-Module-1-Project-Mecaho/data/raw/db_raw.csv')
df.head()
```

# Data cleaning

```python
df.info(memory_usage='deep')
```

### Nulls:

```python
def del_null(df):

    null_cols = df.isnull().sum()/len(df)*100
    null_cols[null_cols>0]
    threshold = 40
    null_filter = null_cols > threshold
    drop_cols = list(null_cols[null_filter].index)
    df.drop(drop_cols, axis=1, inplace=True)

#Apply to:
del_null(df)
df.head()
```

### Replace false null values for 'Nan' :

```python
def re_fnulls(df, col, old_v, new_v):
    df[col] = df[col].replace(old_v, new_v)
    return df

#Apply to:
re_fnulls(df, 'country', 'None', None)
re_fnulls(df, 'age', 'nan', 9999)
df.head()
```

### Delete duplicate columns:

```python
df = df.loc[:,~df.columns.duplicated()]
df.head()
```

### Dropping columns

```python
def drop_c(df, x):
        df.drop([x], axis=1, inplace=True)
        return df
    
#Apply to:
drop_c(df, 'Unnamed: 0')
drop_c(df, 'Unnamed: 0.1')
drop_c(df, 'Unnamed: 0.2')
drop_c(df, 'id.1')
drop_c(df, 'id.2')
drop_c(df, 'realTimePosition')
df.head()
```

### Rename columns

```python
def rename_c(df, old, new):
    df.rename(columns={old: new}, inplace=True)
    return df
```

### Capitalize

```python
def title(df, x):
    df[x] = df[x].str.title()
    return df

title(df,'name')
title(df,'lastName')
df.head()
```

### Cleaning some strings    

```python
def del_str_patt(df, col, del_str, new_str):
    df[col] = df[col].apply(lambda x: re.sub(del_str, new_str, str(x)))
    return df

del_str_patt(df, 'age', '\s[^\d]+', '')
del_str_patt(df, 'worth', ' BUSD', '')
del_str_patt(df, 'worthChange', ' millions USD', '')
df.head()
```

### Slip columns:

```python
def split(df, new_c2, old_c1, x):
    df[new_c2] = df[old_c1].str.split(x).str[1]
    df[old_c1] = df[old_c1].str.split(x).str[0]
    return df

#Apply to:
split(df, 'Company', 'Source', '==>')
df.head()
```
