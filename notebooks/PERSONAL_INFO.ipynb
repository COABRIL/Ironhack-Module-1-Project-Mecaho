```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine
```

```python
sqlitedb_path = '/home/cristina/ironhack/Ironhack-Module-1-Project-Mecaho/data/raw/cortega.db'
engine = create_engine(f'sqlite:///{sqlitedb_path}')
personal_info = pd.read_sql_query("select * from personal_info", engine)
display(personal_info)
```

```python
personal_info['lastName'] = personal_info['lastName'].str.capitalize()
display(personal_info)
```

```python
#Delete the column 'Unnamed:0':
after = len(personal_info.columns)
personal_info = personal_info.drop(['Unnamed: 0'], axis=1)
before = len(personal_info.columns)
print(f'Number of columns dropped: {after - before}')
display(personal_info)
```

```python
personal_info['gender'].value_counts()
```

```python
personal_info['gender'] = personal_info['gender'].str.replace('M', 'Male')
personal_info['gender'] = personal_info['gender'].str.replace('F', 'Female')
display(personal_info)
```
